```
import matplotlib.pyplot as plt
import matplotlib.patches as patches

import matplotlib.collections as mc
import matplotlib.cm as cm

fig = plt.figure()
ax = plt.axes()

# fc=face color ec = edge color
s1 = patches.Circle(xy = (0.3,0.6),radius = 0.3,fill = False,ec = "r")
s2 = patches.Circle(xy = (0.5,0.3),radius = 0.3,fill = False,ec = "r")
s3 = patches.Circle(xy = (0.7,0.7),radius = 0.3,fill = False,ec = "r")
mid1 = patches.Circle(xy = (0.3,0.6),radius = 0.01,fill = True,fc="r",ec = "r")
mid2 = patches.Circle(xy = (0.5,0.3),radius = 0.01,fill = True,fc="r",ec = "r")
mid3 = patches.Circle(xy = (0.7,0.7),radius = 0.01,fill = True,fc="r",ec = "r")


l1 = mc.LineCollection([[(0.199988456, 0.882838631),(0.686561795, 0.534935516)]], linewidths=1,ec="g")
l2 = mc.LineCollection([[(0.686561795, 0.534935516),(0.981068264, 0.804883893)]], linewidths=1,ec="g")

l3 = mc.LineCollection([[(0.044482402, 0.757196556),(0.675845346,0.543060515 )]], linewidths=1,ec="b")
l4 = mc.LineCollection([[(0.675845346,0.543060515 ),(0.584769942,0.976987425 )]], linewidths=1,ec="b")

l5 = mc.LineCollection([[(0.088189676,0.812453258 ),(0.77736711,0.414313107 )]], linewidths=1,ec="c")
l6 = mc.LineCollection([[(0.77736711, 0.414313107),(0.892572819, 0.930034148)]], linewidths=1,ec="c")

l7 = mc.LineCollection([[(0.017061721, 0.699729286),(0.692934383, 0.529730982)]], linewidths=1,ec="m")
l8 = mc.LineCollection([[(0.692934383, 0.529730982),(0.637690632, 0.993457906)]], linewidths=1,ec="m")

l9 = mc.LineCollection([[(0.178823283,0.874437977 ),(0.459467512, 0.597249251)]], linewidths=1,ec="y")
l10 = mc.LineCollection([[(0.459467512, 0.597249251),(0.444516875, 0.857252576)]], linewidths=1,ec="y")



ax.add_patch(s1)
ax.add_patch(s2)
ax.add_patch(s3)
ax.add_patch(mid1)
ax.add_patch(mid2)
ax.add_patch(mid3)

ax.add_collection(l1)
ax.add_collection(l2)
ax.add_collection(l3)
ax.add_collection(l4)
ax.add_collection(l5)
ax.add_collection(l6)
ax.add_collection(l7)
ax.add_collection(l8)
ax.add_collection(l9)
ax.add_collection(l10)

plt.axis("scaled")
ax.set_aspect("equal")
```
